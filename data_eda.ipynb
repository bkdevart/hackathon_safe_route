```python
arrests.columns = ['OBJECTID', 'X', 'Y', 'arre_id', 'case_id', 'agency', 'date_arr',
       'time_arr', 'datetime_arr', 'MONTH_ARR', 'YEAR_ARR', 'DOW_ARR',
       'TIME_ARRST', 'age', 'race', 'sex', 'ethnicity', 'arr_type',
       'neighborhd', 'ADDRESS_PUBLIC', 'city', 'state', 'zip', 'arr_chrg',
       'chrgdesc', 'chrg_cnt', 'fel_misd', 'chrg_seq', 'APPSTATE',
       'LOC_STATUS', 'WARD', 'NHA_NAME', 'TMSECT', 'DIVISION', 'DIVISION_NO',
       'DIVSECT', 'TRSQ', 'City_geo', 'ADDRESS_100BLK']
```

```python
arrests['x_y'] = arrests['X'].astype(str) + ', ' + arrests['Y'].astype(str)
arrests['x_y'].head()
```

```python
crime_list = ['BURGLARY',
 'HOMICIDE',
 'ROBBERY',
 'ASSAULT, AGGRAVATED',
 'LARCENY',
 'ARSON',
 'GTA',
 'SEXUAL ASSAULT']
```

```python
# Crime list
crime_list = ['BURGLARY', 'HOMICIDE', 'ROBBERY', 'ASSAULT, AGGRAVATED', 'LARCENY', 'ARSON', 'GTA', 'SEXUAL ASSAULT']

# Function to map crimes to grouped categories
def map_crime(chrgdesc):
    for crime in crime_list:
        if pd.notna(chrgdesc) and crime.lower() in chrgdesc.lower():
            return crime
    return 'OTHER'  # Return 'OTHER' if no match is found

# Create a new column 'grouped_crime' that maps the matching crime type or 'OTHER'
arrests['grouped_crime'] = arrests['chrgdesc'].apply(map_crime)

# Display the updated DataFrame
arrests[['chrgdesc', 'grouped_crime']].head()
# len(arrests)
```

```python
severity = {
    'BURGLARY': 5, 
    'HOMICIDE': 1, 
    'ROBBERY': 3, 
    'ASSAULT, AGGRAVATED': 4, 
    'LARCENY': 6, 
    'ARSON': 8, 
    'GTA': 5, 
    'SEXUAL ASSAULT': 2
}

coordinates_weight = arrests.copy()
coordinates_weight['severity'] = coordinates_weight['grouped_crime'].map(severity)
```

```python
arrest_counts = pd.DataFrame(arrests.groupby(['x_y', 'grouped_crime'])['OBJECTID'].count()).reset_index()
arrest_counts.head()
```

```python
print(len(arrest_counts))
coordinates_weight[['X', 'Y', 'chrgdesc', 'grouped_crime']].to_csv('data/coordinates_weight.csv')
```
