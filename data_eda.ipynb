```python
# Crime list
crime_list = ['BURGLARY', 'HOMICIDE', 'ROBBERY', 'ASSAULT, AGGRAVATED', 'LARCENY', 'ARSON', 'GTA', 'SEXUAL ASSAULT']

# Function to map crimes to grouped categories
def map_crime(chrgdesc):
    for crime in crime_list:
        if pd.notna(chrgdesc) and crime.lower() in chrgdesc.lower():
            return crime
    return 'OTHER'  # Return 'OTHER' if no match is found

# Create a new column 'grouped_crime' that maps the matching crime type or 'OTHER'
arrests['grouped_crime'] = arrests['chrgdesc'].apply(map_crime)

# Display the updated DataFrame
arrests[['chrgdesc', 'grouped_crime']].head()
# len(arrests)
```

```python
severity = {
    'BURGLARY': 5, 
    'HOMICIDE': 1, 
    'ROBBERY': 3, 
    'ASSAULT, AGGRAVATED': 4, 
    'LARCENY': 6, 
    'ARSON': 8, 
    'GTA': 5, 
    'SEXUAL ASSAULT': 2
}

coordinates_weight = arrests.copy()
coordinates_weight['severity'] = coordinates_weight['grouped_crime'].map(severity)
```

```python
# TODO: create some api requests to pull lat/lon from address in crimes data
import requests
from config import gmaps_key as api_key


def get_lat_lon(address, api_key):
    # URL for the Google Maps Geocoding API
    geocode_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    # Make the API request with the address and API key
    params = {
        'address': address,
        'key': api_key
    }
    
    # Send the request to the geocoding API
    response = requests.get(geocode_url, params=params)
    
    # Parse the response JSON
    result = response.json()
    
    # Extract latitude and longitude from the response
    if result['status'] == 'OK':
        lat = result['results'][0]['geometry']['location']['lat']
        lon = result['results'][0]['geometry']['location']['lng']
        return lat, lon
    else:
        return None, None

# Example usage:
address = "1600 Amphitheatre Parkway, Mountain View, CA"
lat, lon = get_lat_lon(address, api_key)

print(f"Latitude: {lat}, Longitude: {lon}")
```

```python
from sklearn.cluster import KMeans
X = coordinates_weight['chrgdesc'].values.reshape(-1, 1)
# kmeans = KMeans(n_clusters=8, random_state=0)
# crimes['cluster'] = kmeans.fit_predict(X)
# crimes.head()
from time import time

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_df=0.5,
    min_df=5,
    stop_words="english",
)
t0 = time()
X_tfidf = vectorizer.fit_transform(coordinates_weight['chrgdesc'])
print(X_tfidf)
```

```python
kmeans = KMeans(n_clusters=8, random_state=42)
# fit the model
kmeans.fit(X_tfidf)
# store cluster labels in a variable
clusters = kmeans.labels_
```

```python
coordinates_weight['cluster'] = clusters
coordinates_weight[['chrgdesc', 'cluster']].head()
```

```python
from sklearn.decomposition import PCA

# initialize PCA with 2 components
pca = PCA(n_components=2, random_state=42)
# pass our X to the pca and store the reduced vectors into pca_vecs
pca_vecs = pca.fit_transform(X_tfidf.toarray())
# save our two dimensions into x0 and x1
x0 = pca_vecs[:, 0]
x1 = pca_vecs[:, 1]
coordinates_weight['pca1'] = x0
coordinates_weight['pca2'] = x1
```
